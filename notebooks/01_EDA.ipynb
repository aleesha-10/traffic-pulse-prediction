```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# optional for warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load your CSV file
file_path = "C:\\Users\\Laptop\\Desktop\\traffic-pulse-prediction\\data\\pems-8w.csv"  # update with actual file name
df = pd.read_csv(file_path, header=None)

# Quick look at the data
df.head()
df.info()
df.describe()
```

```python
# converting everything to numeric
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
# handling missing values
df = df.fillna(method="ffill").fillna(method="bfill")
# or df = df.dropna()
```

```python
# adding sensor names
df.columns = [f"sensor_{i}" for i in range(df.shape[1])]

# adding time index
df.index = pd.date_range(
    start="2017-01-01",
    periods=len(df),
    freq="5min"
)
df.index.name = "timestamp"

```

```python
# Create your modeling signal
traffic = df.mean(axis=1).to_frame(name="traffic_flow")
traffic.head()
```

```python
import os

os.makedirs("../data/processed", exist_ok=True)
traffic.to_csv("../data/processed/traffic_aggregated.csv")
```
